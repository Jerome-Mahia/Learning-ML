**Prepare the data**

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
data = pd.DataFrame({'Mood': ['Dark', 'Semi-Dark','Bright', 'Semi-Dark','Bright','Bright','Bright'],
 'Rating': ['Adult', 'Adult', 'Family','Adult','Family','Adult', 'Adult'],
 'Class': ['Dislike','Dislike','Like','Dislike', 'Like','Like','Like']
 })
data
```

```python
data[['Mood','Rating']]
```

```python
X = pd.get_dummies(data[['Mood','Rating']]) # Hot encoding the dataset
X.head(2)
```

```python
encoder = LabelEncoder()
```

```python
y = encoder.fit_transform(data['Class'])
```

**Build Model**

```python
model = Sequential([
    Dense(16,input_dim=5),
    Dense(8),
    Dense(2)
])
model.summary()
```

**Train the model**

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, y, epochs=3)
```
